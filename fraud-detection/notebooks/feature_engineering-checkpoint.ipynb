```python
# Feature Engineering

## Objective
This notebook creates additional features from the raw transaction data
to improve fraud detection performance. These features capture temporal,
behavioral, and geographic patterns associated with fraudulent activity.
```

```python
import pandas as pd
import numpy as np

fraud_df = pd.read_csv("../data/raw/Fraud_Data.csv")
```

```python
# Convert strings to actual Date objects
fraud_df['signup_time'] = pd.to_datetime(fraud_df['signup_time'])
fraud_df['purchase_time'] = pd.to_datetime(fraud_df['purchase_time'])

# Create new number columns
fraud_df['hour_of_day'] = fraud_df['purchase_time'].dt.hour
fraud_df['day_of_week'] = fraud_df['purchase_time'].dt.dayofweek

# Calculate the time difference in seconds (very useful for fraud!)
fraud_df['time_diff'] = (fraud_df['purchase_time'] - fraud_df['signup_time']).dt.total_seconds()

# Now we can drop the original date strings
fraud_df = fraud_df.drop(columns=['signup_time', 'purchase_time'])
```

```python
# Select the columns with words
cat_cols = ['source', 'browser', 'sex']

# Use get_dummies to create 0/1 columns automatically
fraud_df = pd.get_dummies(fraud_df, columns=cat_cols)
```

```python
transaction_counts = fraud_df.groupby('user_id').size()
fraud_df['user_transaction_count'] = fraud_df['user_id'].map(transaction_counts)
```

```python
### Transaction Frequency

Users with unusually high numbers of transactions may indicate
automated or suspicious behavior. This feature captures transaction
frequency at the user level.
```

```python
# Drop columns that are just random IDs
fraud_df = fraud_df.drop(columns=['user_id', 'device_id', 'ip_address'], errors='ignore')
```

```python
# This shows any columns that are NOT numbers
print(fraud_df.select_dtypes(exclude=['number']).columns)
```

```python
# Convert True/False to 1/0
fraud_df = fraud_df.astype(int)
fraud_df.info()
```

```python
# A simple way to balance classes if SMOTE is broken
# Separate fraud and non-fraud
fraud = fraud_df[fraud_df['class'] == 1]
non_fraud = fraud_df[fraud_df['class'] == 0]

# Take a small random sample of non-fraud to match the number of fraud cases
non_fraud_downsampled = non_fraud.sample(n=len(fraud), random_state=42)

# Combine them back together
balanced_df = pd.concat([fraud, non_fraud_downsampled])

print(f"New distribution: {balanced_df['class'].value_counts()}")
```

```python
I chose undersampling to create a balanced training set where fraud
and legitimate cases are equal, ensuring the model treats both classes with equal importance
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Plot before undersampling
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='class', data=fraud_df)
plt.title('Before Undersampling (Imbalanced)')

# 2. Plot after undersampling
plt.subplot(1, 2, 2)
sns.countplot(x='class', data=balanced_df)
plt.title('After Undersampling (Balanced)')

plt.tight_layout()
plt.show()
```
